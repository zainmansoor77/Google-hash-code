# BUILT-IN command for ascending order

```python
a = [9,8,6,7,1,5,3,4,6,9,23,54,11,8,98]
b = a.copy()
b.sort()
print(b)
print(a)
```

```python
a = [9,8,6,7,1,5,3,4,6,9,23,54,11,8,98]
b = a
b.sort()
print(b)
print(a)
```

# Bubble Sort Algorithm for ascending arrangement

```python
a = [100,9,8,6,7,1,5,104,3,4,6,9,23,54,11,8,98]
t = len(a)
for j in range(t ,0,-1):
    for i in range(1,j):
        if a[i-1] > a[i]:
            a[i - 1],a[i] = a[i], a[i - 1]
     
print(a)    
```

# BUILT-IN command for descending order

```python
a = [100,9,8,6,7,1,5,104,3,4,6,9,23,54,11,8,98]
a.sort(reverse = True)
print(a)
```

# Bubble Sort Algorithm for descending arrangement

```python
a = [100,9,8,6,7,1,5,104,3,4,6,9,23,54,11,8,98]
t = len(a)

for j in range(t-1,0,-1):
    for i in range(j):
        if a[i] < a[i +1]:
            a[i],a[i + 1] = a[i + 1], a[i]
        
print(a)        
```
